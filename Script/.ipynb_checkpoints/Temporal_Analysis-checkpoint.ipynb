```python
def compute_avg_top_dist_gvn_dt_rand(df,Ld_dic,rand_list,dt_list,seed_list,title1):

    df.timestamp = df.timestamp - min(df.timestamp)

    #sort the timestamps and identify the precision of the dataset (dt)





    df = df.sort_values('timestamp',ascending=True).reset_index().drop('index',axis=1)

    # remove doubles (optional)
    df.drop_duplicates(['nodes','timestamp'],inplace = True)
    for seed in seed_list:
    ### general dictionary
        tot_dic_dst_counter_gen = {}

        # In[7]:
        # charge the original df, set strings in tuples and normalize to the min


        



        # introduce the two main dictionaries

        tot_dic_dst_counter1 = {}
        tot_dic_dst_counter2 = {}
        tot_dic_dst_counter3 = {}







        df1 = shuffle_df(df,seed)     ### permute timeseries (IT SHOULD BE RAND2!!!!)
        df2 = permute_timestamps(df,seed)     ### permute timestamp
        df3 = random_df_same_weight(df,seed)  ### permute timeseries and preserve number of contacts per link


        df1.sort_values('timestamp',ascending = True,inplace = True)
        df2.sort_values('timestamp',ascending = True,inplace = True)
        df3.sort_values('timestamp',ascending = True,inplace = True)

        ###---------------- Check  total number of contacts and distribution of number of contacts per link is preserved-------------------###
        assert Counter(df.groupby(df.nodes).size().values) == Counter(df1.groupby(df1.nodes).size().values)
        assert Counter(df.groupby(df.nodes).size().values) == Counter(df2.groupby(df2.nodes).size().values)
        assert Counter(df.groupby(df.nodes).size().values) == Counter(df3.groupby(df3.nodes).size().values)
        assert Counter(df2.groupby(df2.nodes).size().values) == Counter(df3.groupby(df3.nodes).size().values)




        t_inf =  (np.max(df.timestamp) - np.min(df.timestamp))  + 100   ### introduce t_inf to check convergence to 1 of E[tau]

        del df



        ###----------------- Mean contact average distance ------------------------###

        distance_rel = Ld_dic



        cnt_ev1 = get_mean_distance_events(Ld_dic,df1)
        cnt_ev2 = get_mean_distance_events(Ld_dic,df2)
        cnt_ev3 = get_mean_distance_events(Ld_dic,df3)



        tot_dic_dst_counter_gen['mean_distance_rand1'] = DescrStatsW(cnt_ev1.index,weights = cnt_ev1.values).mean


        tot_dic_dst_counter_gen['mean_distance_rand2'] = DescrStatsW(cnt_ev2.index,weights = cnt_ev2.values).mean

        tot_dic_dst_counter_gen['mean_distance_rand3'] = DescrStatsW(cnt_ev3.index,weights = cnt_ev3.values).mean

        ###----------------- Check that the total number of pair of contacts is the total number of possible pairs of a set of  df.shape[0] elements ------------------------###
        n5 = df1.shape[0]
        n6 = df2.shape[0]
        n7 = df3.shape[0]
        assert (n5*(n5-1))/2 == sum(cnt_ev1.values)
        assert  (n6*(n6-1))/2 == sum(cnt_ev2.values)
        assert (n7*(n7-1))/2 == sum(cnt_ev3.values)



        dt_list = dt_list+[t_inf]

        for t in  dt_list:

            cnt_dst1 = Counter()
            cnt_spk1 = Counter()
            cnt_dst2 = Counter()
            cnt_spk2 = Counter()
            cnt_dst3 = Counter()
            cnt_spk3 = Counter()

            max_time = max(df1.timestamp)
            index_set = set()
            a = datetime.datetime.now()
            t_cnt = 0
            e_cnt = 0
            i = 0

            n0_1 = 0
            n_tot1 = 0

            n0_2 = 0
            n_tot2 = 0

            n0_3 = 0
            n_tot3 = 0

            for ref_time in tqdm(df1.timestamp.unique()):

                #####------------------------------------------------------- OVERLAPP CASE------------------------------------------------------###
                ###All possible combinations of contacts in the time window, some pairs are overcounted, it ends when the time window is larger than maximum time

                delta_t = t
                t_cnt+=1


            #####------------------------------------------------------- NO OVERLAPP CASE------------------------------------------------------###
            ###All pairs of contacts that contains  the original contact , no pairs are overcounted, it ends when the starting timestamp of the contact is larger than the maximum t
                if (ref_time <= max_time):

                    ### Select part of the dataframe where  timestamp < t <timestamp + delta t
                    ### group together initial with timestamp == ref_time
                    ### group  together the other contacts
                    ### ntot measures always the total number of combinations among contacts of df_rest, cosidering that ref_nodes can be paired together, while the others can't, N_couples_df_rest = bin(N_ref_nodes)+(N_ref_nodes*N_other nodes)

                    if 0 in randomization_list:

                        df_rest1 = df1[(df1.timestamp >= ref_time)&(df1.timestamp < ref_time + delta_t)]
                        ref_nodes1 = df_rest1[df_rest1.timestamp == ref_time].nodes
                        df_rest1 = df_rest1[df_rest1.timestamp > ref_time]

                    if 1 in randomization_list:
                        df_rest2 = df2[(df2.timestamp >= ref_time)&(df2.timestamp < ref_time + delta_t)]
                        ref_nodes2 = df_rest2[df_rest2.timestamp == ref_time].nodes
                        df_rest2 = df_rest2[df_rest2.timestamp > ref_time]

                    if 2 in randomization_list:
                        df_rest3 = df3[(df3.timestamp >= ref_time)&(df3.timestamp < ref_time + delta_t)]
                        ref_nodes3 = df_rest3[df_rest3.timestamp == ref_time].nodes
                        df_rest3 = df_rest3[df_rest3.timestamp > ref_time]

                    ### if delta t is small use  itertools combinations


                    if t < 640:
                        if 0 in randomization_list:
                            size_n1 = df_rest1.shape[0]
                            n_tot1 += ref_nodes1.shape[0] *(size_n1)
                            n_tot1 += ((ref_nodes1.shape[0]*(ref_nodes1.shape[0]-1)))/2.

                            for x,y in itertools.combinations(ref_nodes1,2):
                                cnt_dst1.update({Ld_dic[x,y]:1})
                            for x,y in itertools.product(ref_nodes1,df_rest1.nodes):
                                cnt_dst1.update({Ld_dic[x,y]:1})


                        if 1 in randomization_list:
                            size_n2 = df_rest2.shape[0]
                            n_tot2 += ref_nodes2.shape[0] *(size_n2)
                            n_tot2 += ((ref_nodes2.shape[0]*(ref_nodes2.shape[0]-1)))/2.

                            for x,y in itertools.combinations(ref_nodes2,2):
                                cnt_dst2.update({Ld_dic[x,y]:1})
                            for x,y in itertools.product(ref_nodes2,df_rest2.nodes):
                                cnt_dst2.update({Ld_dic[x,y]:1})

                        if 2 in randomization_list:
                            size_n3 = df_rest3.shape[0]
                            n_tot3 += ref_nodes3.shape[0] *(size_n3)
                            n_tot3 += ((ref_nodes3.shape[0]*(ref_nodes3.shape[0]-1)))/2.

                            for x,y in itertools.combinations(ref_nodes3,2):
                                cnt_dst3.update({Ld_dic[x,y]:1})
                            for x,y in itertools.product(ref_nodes3,df_rest3.nodes):
                                cnt_dst3.update({Ld_dic[x,y]:1})


                    ### if delta t is larger compute distance onces and then use multiplicity of contacts
                    else:
                        if 0 in randomization_list:
                            size_n1 = df_rest1.groupby(df_rest1.nodes).size()
                            size1 = size_n1.to_dict()
                            n_tot1 += ref_nodes1.shape[0] *(sum(size_n1.values))
                            n_tot1 += ((ref_nodes1.shape[0]*(ref_nodes1.shape[0]-1)))/2.
                            for x,y in itertools.combinations(ref_nodes1,2):
                                cnt_dst1.update({Ld_dic[x,y]:1})

                            for x,y in itertools.product(ref_nodes1,df_rest1.nodes.unique()):
                                cnt_dst1.update({Ld_dic[x,y]:size1[y]})



                        if 1 in randomization_list:
                            size_n2 = df_rest2.groupby(df_rest2.nodes).size()
                            size2 = size_n2.to_dict()
                            n_tot2 += ref_nodes2.shape[0] *(sum(size_n2.values))
                            n_tot2 += ((ref_nodes2.shape[0]*(ref_nodes2.shape[0]-1)))/2.
                            for x,y in itertools.combinations(ref_nodes2,2):
                                cnt_dst2.update({Ld_dic[x,y]:1})

                            for x,y in itertools.product(ref_nodes2,df_rest2.nodes.unique()):
                                cnt_dst2.update({Ld_dic[x,y]:size2[y]})


                        if 2 in randomization_list:
                            size_n3 = df_rest3.groupby(df_rest3.nodes).size()
                            size3 = size_n3.to_dict()
                            n_tot3 += ref_nodes3.shape[0] *(sum(size_n3.values))
                            n_tot3 += ((ref_nodes3.shape[0]*(ref_nodes3.shape[0]-1)))/2.
                            for x,y in itertools.combinations(ref_nodes3,2):
                                cnt_dst3.update({Ld_dic[x,y]:1})

                            for x,y in itertools.product(ref_nodes3,df_rest3.nodes.unique()):
                                cnt_dst3.update({Ld_dic[x,y]:size3[y]})





                    e_cnt+=1
                    i+=1
        ###-------------------------------------------------------------------- FIRST CONSISTENCY CHECK -------------------------------------------------------------####
        ### Check that the sum of N_pairs_df == total number of values of the counter
            if 0 in randomization_list: assert sum(cnt_dst1.values())==n_tot1
            if 1 in randomization_list: assert sum(cnt_dst2.values())==n_tot2
            if 2 in randomization_list: assert sum(cnt_dst3.values())==n_tot3

        ###-------------------------------------------------------------------- CONVERGENCE CHECK -------------------------------------------------------------####
        ### Check that if delta t------> inf E[eta|delta t] == E[eta] and the distribution is the same
            if delta_t == max_time+100:
                if 0 in randomization_list:
                    assert DescrStatsW(cnt_dst1.keys(),cnt_dst1.values()).mean==DescrStatsW(cnt_ev1.index,weights = cnt_ev1.values).mean
                    assert cnt_dst1 == cnt_ev1.to_dict()
                if 1 in randomization_list:
                    assert DescrStatsW(cnt_dst2.keys(),cnt_dst2.values()).mean==DescrStatsW(cnt_ev2.index,weights = cnt_ev2.values).mean
                    assert cnt_dst2 == cnt_ev2.to_dict()
                if 2 in randomization_list:
                    assert DescrStatsW(cnt_dst3.keys(),cnt_dst3.values()).mean==DescrStatsW(cnt_ev3.index,weights = cnt_ev3.values).mean
                    assert cnt_dst3 == cnt_ev3.to_dict()

                break

            if 0 in randomization_list:
                tot_dic_dst_counter1[t] = cnt_dst1
            if 1 in randomization_list:
                tot_dic_dst_counter2[t] = cnt_dst2
            if 2 in randomization_list:
                tot_dic_dst_counter3[t] = cnt_dst3

        ###-------------------------------------------------------------------- SECOND CONSISTENCY CHECK -------------------------------------------------------------####
        ### Check that the total number of df_rest analysed is the same as the total number of unique timestamp of df
            del cnt_dst1,cnt_dst2,cnt_dst3
            assert e_cnt==t_cnt
            assert t_cnt==df1.timestamp.unique().shape[0]
            print str(t)+" completed"
            
        tot_dic_dst_counter_gen['distances_rand1'] = tot_dic_dst_counter1
        tot_dic_dst_counter_gen['distances_rand2'] = tot_dic_dst_counter2
        tot_dic_dst_counter_gen['distances_rand3'] = tot_dic_dst_counter3


        try: os.mkdir('../Results/avg_top_dist_gvn_temp/'+title1+str1+'/')
        except: print '../Results/avg_top_dist_gvn_temp/'+title1+str1+'/ already exist'



        #joblib.dump(tot_dic_spk_counter,'/tudelft.net/staff-bulk/ewi/insy/MMC/aceria/Temporal_Network/Other_Networks/statistical_analysis/'+title+'_rand2_clean/'+'spk_tot_stat_new2_'+title1+'_rand2_'+'seed'+str(seed)+'.joblib')



        #joblib.dump(tot_dic_dst_counter_gen,'../Results/avg_top_dist_gvn_temp/'+title1+str1+'/'+title1+'_rand_'+'seed'+str(seed)+'.joblib')

```

```python
def compute_avg_top_dist_gvn_dt_rand(df, Ld_dic, rand_list, dt_list, seed_list, title1):
     """
    Compute average topological distances in a temporal network with various randomizations.
    
    Args:
        df (DataFrame): The input DataFrame containing network data.
        Ld_dic (dict): A dictionary used for calculating distances between nodes.
        rand_list (list): A list indicating which randomization methods to apply (0, 1, 2, or a combination).
        dt_list (list): A list of time windows (delta_t) for calculating distances.
        seed_list (list): A list of randomization seed values.
        title1 (str): A title or identifier for the results.
    """
    
    
    
    # Adjust timestamps to start from zero
    df.timestamp = df.timestamp - min(df.timestamp)

    # Sort the DataFrame by timestamps in ascending order
    df = df.sort_values('timestamp', ascending=True).reset_index().drop('index', axis=1)

    # Remove duplicate rows with the same 'nodes' and 'timestamp'
    df.drop_duplicates(['nodes', 'timestamp'], inplace=True)

    # Iterate over different randomization seeds
    for seed in seed_list:
        # Initialize dictionaries to store distance-related information
        tot_dic_dst_counter_gen = {}  # General dictionary for distance results
        tot_dic_dst_counter1 = {}     # Dictionary for randomization 1 distance results
        tot_dic_dst_counter2 = {}     # Dictionary for randomization 2 distance results
        tot_dic_dst_counter3 = {}     # Dictionary for randomization 3 distance results

        # Perform three different randomizations on the input data
        df1 = shuffle_df(df, seed)        # Randomization 1: Permutes timeseries
        df2 = permute_timestamps(df, seed)  # Randomization 2: Permutes timestamps
        df3 = random_df_same_weight(df, seed)  # Randomization 3: Permutes timeseries, preserving contact weights

        # Sort the randomized DataFrames by timestamp
        df1.sort_values('timestamp', ascending=True, inplace=True)
        df2.sort_values('timestamp', ascending=True, inplace=True)
        df3.sort_values('timestamp', ascending=True, inplace=True)

        # Check that the total number of contacts and distribution of contacts per link are preserved
        assert Counter(df.groupby(df.nodes).size().values) == Counter(df1.groupby(df1.nodes).size().values)
        assert Counter(df.groupby(df.nodes).size().values) == Counter(df2.groupby(df2.nodes).size().values)
        assert Counter(df.groupby(df.nodes).size().values) == Counter(df3.groupby(df3.nodes).size().values)
        assert Counter(df2.groupby(df2.nodes).size().values) == Counter(df3.groupby(df3.nodes).size().values)

        # Calculate t_inf, which is used for convergence checks
        t_inf = (np.max(df.timestamp) - np.min(df.timestamp)) + 100

        # Calculate mean contact average distance for each randomization
        cnt_ev1 = get_mean_distance_events(Ld_dic, df1)
        cnt_ev2 = get_mean_distance_events(Ld_dic, df2)
        cnt_ev3 = get_mean_distance_events(Ld_dic, df3)

        # Store the mean contact average distances in the general dictionary
        tot_dic_dst_counter_gen['mean_distance_rand1'] = DescrStatsW(cnt_ev1.index, weights=cnt_ev1.values).mean
        tot_dic_dst_counter_gen['mean_distance_rand2'] = DescrStatsW(cnt_ev2.index, weights=cnt_ev2.values).mean
        tot_dic_dst_counter_gen['mean_distance_rand3'] = DescrStatsW(cnt_ev3.index, weights=cnt_ev3.values).mean

        # Check that the total number of pairs of contacts matches expectations
        n5 = df1.shape[0]
        n6 = df2.shape[0]
        n7 = df3.shape[0]
        assert (n5 * (n5 - 1)) / 2 == sum(cnt_ev1.values)
        assert (n6 * (n6 - 1)) / 2 == sum(cnt_ev2.values)
        assert (n7 * (n7 - 1)) / 2 == sum(cnt_ev3.values)

        # Append t_inf to the dt_list
        dt_list = dt_list + [t_inf]

        # Iterate over different time windows (t)
        for t in dt_list:
            # Initialize counters for distances and statistics
            cnt_dst1 = Counter()
            cnt_spk1 = Counter()
            cnt_dst2 = Counter()
            cnt_spk2 = Counter()
            cnt_dst3 = Counter()
            cnt_spk3 = Counter()

            # Find the maximum timestamp in the original DataFrame
            max_time = max(df1.timestamp)
            index_set = set()
            a = datetime.datetime.now()
            t_cnt = 0
            e_cnt = 0
            i = 0

            n0_1 = 0
            n_tot1 = 0

            n0_2 = 0
            n_tot2 = 0

            n0_3 = 0
            n_tot3 = 0

            # Iterate over unique timestamps in the DataFrame
            for ref_time in tqdm(df1.timestamp.unique()):
                # Perform calculations based on the time window (delta_t)
                delta_t = t
                t_cnt += 1

                # Handle the OVERLAP CASE (All possible combinations of contacts in the time window)
                # Handle the NO OVERLAP CASE (All pairs of contacts containing the original contact)
                if (ref_time <= max_time):
                    # Select part of the dataframe where timestamp < t < timestamp + delta t
                    # Group together initial contacts with timestamp == ref_time
                    # Group together the other contacts
                    # ntot measures the total number of combinations among contacts of df_rest,
                    # considering that ref_nodes can be paired together, while the others can't
                    # N_couples_df_rest = bin(N_ref_nodes) + (N_ref_nodes * N_other nodes)

                    if 0 in randomization_list:
                        df_rest1 = df1[(df1.timestamp >= ref_time) & (df1.timestamp < ref_time + delta_t)]
                        ref_nodes1 = df_rest1[df_rest1.timestamp == ref_time].nodes
                        df_rest1 = df_rest1[df_rest1.timestamp > ref_time]

                    if 1 in randomization_list:
                        df_rest2 = df2[(df2.timestamp >= ref_time) & (df2.timestamp < ref_time + delta_t)]
                        ref_nodes2 = df_rest2[df_rest2.timestamp == ref_time].nodes
                        df_rest2 = df_rest2[df_rest2.timestamp > ref_time]

                    if 2 in randomization_list:
                        df_rest3 = df3[(df3.timestamp >= ref_time) & (df3.timestamp < ref_time + delta_t)]
                        ref_nodes3 = df_rest3[df_rest3.timestamp == ref_time].nodes
                        df_rest3 = df_rest3[df_rest3.timestamp > ref_time]

                    # If delta_t is small, use itertools combinations
                    if t < 640:
                        if 0 in randomization_list:
                            size_n1 = df_rest1.shape[0]
                            n_tot1 += ref_nodes1.shape[0] * (size_n1)
                            n_tot1 += ((ref_nodes1.shape[0] * (ref_nodes1.shape[0] - 1))) / 2.
                            for x, y in itertools.combinations(ref_nodes1, 2):
                                cnt_dst1.update({Ld_dic[x, y]: 1})
                            for x, y in itertools.product(ref_nodes1, df_rest1.nodes):
                                cnt_dst1.update({Ld_dic[x, y]: 1})

                        if 1 in randomization_list:
                            size_n2 = df_rest2.shape[0]
                            n_tot2 += ref_nodes2.shape[0] * (size_n2)
                            n_tot2 += ((ref_nodes2.shape[0] * (ref_nodes2.shape[0] - 1))) / 2.
                            for x, y in itertools.combinations(ref_nodes2, 2):
                                cnt_dst2.update({Ld_dic[x, y]: 1})
                            for x, y in itertools.product(ref_nodes2, df_rest2.nodes):
                                cnt_dst2.update({Ld_dic[x, y]: 1})

                        if 2 in randomization_list:
                            size_n3 = df_rest3.shape[0]
                            n_tot3 += ref_nodes3.shape[0] * (size_n3)
                            n_tot3 += ((ref_nodes3.shape[0] * (ref_nodes3.shape[0] - 1))) / 2.
                            for x, y in itertools.combinations(ref_nodes3, 2):
                                cnt_dst3.update({Ld_dic[x, y]: 1})
                            for x, y in itertools.product(ref_nodes3, df_rest3.nodes):
                                cnt_dst3.update({Ld_dic[x, y]: 1})
                    # If delta_t is larger, compute distance once and then use multiplicity of contacts
                    else:
                        if 0 in randomization_list:
                            size_n1 = df_rest1.groupby(df_rest1.nodes).size()
                            size1 = size_n1.to_dict()
                            n_tot1 += ref_nodes1.shape[0] * (sum(size_n1.values))
                            n_tot1 += ((ref_nodes1.shape[0] * (ref_nodes1.shape[0] - 1))) / 2.
                            for x, y in itertools.combinations(ref_nodes1, 2):
                                cnt_dst1.update({Ld_dic[x, y]: 1})
                            for x, y in itertools.product(ref_nodes1, df_rest1.nodes.unique()):
                                cnt_dst1.update({Ld_dic[x, y]: size1[y]})

                        if 1 in randomization_list:
                            size_n2 = df_rest2.groupby(df_rest2.nodes).size()
                            size2 = size_n2.to_dict()
                            n_tot2 += ref_nodes2.shape[0] * (sum(size_n2.values))
                            n_tot2 += ((ref_nodes2.shape[0] * (ref_nodes2.shape[0] - 1))) / 2.
                            for x, y in itertools.combinations(ref_nodes2, 2):
                                cnt_dst2.update({Ld_dic[x, y]: 1})
                            for x, y in itertools.product(ref_nodes2, df_rest2.nodes.unique()):
                                cnt_dst2.update({Ld_dic[x, y]: size2[y]})

                        if 2 in randomization_list:
                            size_n3 = df_rest3.groupby(df_rest3.nodes).size()
                            size3 = size_n3.to_dict()
                            n_tot3 += ref_nodes3.shape[0] * (sum(size_n3.values))
                            n_tot3 += ((ref_nodes3.shape[0] * (ref_nodes3.shape[0] - 1))) / 2.
                            for x, y in itertools.combinations(ref_nodes3, 2):
                                cnt_dst3.update({Ld_dic[x, y]: 1})
                            for x, y in itertools.product(ref_nodes3, df_rest3.nodes.unique()):
                                cnt_dst3.update({Ld_dic[x, y]: size3[y]})

                    e_cnt += 1
                    i += 1
            # FIRST CONSISTENCY CHECK: Check that the sum of N_pairs_df == total number of values of the counter
            if 0 in randomization_list:
                assert sum(cnt_dst1.values()) == n_tot1
            if 1 in randomization_list:
                assert sum(cnt_dst2.values()) == n_tot2
            if 2 in randomization_list:
                assert sum(cnt_dst3.values()) == n_tot3

            # CONVERGENCE CHECK: Check that if delta t ------> inf E[eta|delta t] == E[eta] and the distribution is the same
            if delta_t == max_time + 100:
                if 0 in randomization_list:
                    assert DescrStatsW(cnt_dst1.keys(), cnt_dst1.values()).mean == DescrStatsW(
                        cnt_ev1.index, weights=cnt_ev1.values).mean
                    assert cnt_dst1 == cnt_ev1.to_dict()
                if 1 in randomization_list:
                    assert DescrStatsW(cnt_dst2.keys(), cnt_dst2.values()).mean == DescrStatsW(
                        cnt_ev2.index, weights=cnt_ev2.values).mean
                    assert cnt_dst2 == cnt_ev2.to_dict()
                if 2 in randomization_list:
                    assert DescrStatsW(cnt_dst3.keys(), cnt_dst3.values()).mean == DescrStatsW(
                        cnt_ev3.index, weights=cnt_ev3.values).mean
                    assert cnt_dst3 == cnt_ev3.to_dict()

                break  # Exit the loop if delta_t is very large (convergence check)

            # Store the distance results in the respective dictionaries
            if 0 in randomization_list:
                tot_dic_dst_counter1[t] = cnt_dst1
            if 1 in randomization_list:
                tot_dic_dst_counter2[t] = cnt_dst2
            if 2 in randomization_list:
                tot_dic_dst_counter3[t] = cnt_dst3

            # SECOND CONSISTENCY CHECK: Check that the total number of df_rest analyzed is the same as the total number of unique timestamps of df
            del cnt_dst1, cnt_dst2, cnt_dst3
            assert e_cnt == t_cnt
            assert t_cnt == df1.timestamp.unique().shape[0]
            print(str(t) + " completed")

        # Store the distance dictionaries for this seed in the general dictionary
        tot_dic_dst_counter_gen['distances_rand1'] = tot_dic_dst_counter1
        tot_dic_dst_counter_gen['distances_rand2'] = tot_dic_dst_counter2
        tot_dic_dst_counter_gen['distances_rand3'] = tot_dic_dst_counter3

        try:
            os.mkdir('../Results/avg_top_dist_gvn_temp/' + title1 + str1 + '/')
        except:
            print('../Results/avg_top_dist_gvn_temp/' + title1 + str1 + '/ already exists')

        # Save the results using joblib
        #joblib.dump(tot_dic_dst_counter_gen,
                    #'../Results/avg_top_dist_gvn_temp/' + title1 + str1 + '/' + title1 + '_rand_' + 'seed' + str(
                     #   seed) + '.joblib')
        return tot_dic_dst_counter_gen
```

```python

def compute_avg_top_dist_gvn_dt(df,Ld_dic,dt_list,title1):

    # In[3]:
    tot_dic_dst_counter_gen = {}

    df.timestamp = df.timestamp - min(df.timestamp)


    # In[7]:
    # remove doubles (optional)


    df = df.sort_values('timestamp',ascending=True).reset_index().drop('index',axis=1)
    df.drop_duplicates(['nodes','timestamp'],inplace = True)


    df.head()
    # introduce the two main dictionaries

    tot_dic_dst_counter = {}


    t_inf =  (np.max(df.timestamp) - np.min(df.timestamp))  + 100   ### introduce t_inf to check convergence to 1 of E[tau]


    ###----------------- Mean contact average distance ------------------------###




    cnt_ev1 = get_mean_distance_events(Ld_dic,df)

    tot_dic_dst_counter_gen['mean_distance'] = DescrStatsW(cnt_ev1.index,weights = cnt_ev1.values).mean
    n5 = df.shape[0]


    ###----------------- Check that the total number of pair of contacts is the total number of possible pairs of a set of  df.shape[0] elements ------------------------###

    assert (n5*(n5-1))/2 == sum(cnt_ev1.values)

    dt_list = dt_list+[t_inf]



    for t in dt_list:


        cnt_dst = Counter()
        max_time = max(df.timestamp)
        index_set = set()
        a = datetime.datetime.now()
        t_cnt = 0
        e_cnt = 0
        i = 0
        n0 = 0
        n_tot = 0

        for ref_time in tqdm(df.timestamp.unique()):



                delta_t = t
                t_cnt+=1


                if (ref_time <= max_time):

                    df_rest = df[(df.timestamp >= ref_time)&(df.timestamp < ref_time + delta_t)]
                    ref_nodes = df_rest[df_rest.timestamp == ref_time].nodes
                    df_rest = df_rest[df_rest.timestamp > ref_time]
                    if t < 640:
                        size_n = df_rest.shape[0]

                        n_tot += ref_nodes.shape[0] *(size_n)
                        n_tot += ((ref_nodes.shape[0]*(ref_nodes.shape[0]-1)))/2.

                        for x,y in itertools.combinations(ref_nodes,2):
                            cnt_dst.update({Ld_dic[x,y]:1})

                        for x,y in itertools.product(ref_nodes,df_rest.nodes):
                            cnt_dst.update({Ld_dic[x,y]:1})
                    else:
                        size_n = df_rest.groupby(df.nodes).size()
                        size = size_n.to_dict()
                        n_tot += ref_nodes.shape[0] *(sum(size_n.values))

                        n_tot += ((ref_nodes.shape[0]*(ref_nodes.shape[0]-1)))/2.
                        for x,y in itertools.combinations(ref_nodes,2):
                            cnt_dst.update({Ld_dic[x,y]:1})

                        for x,y in itertools.product(ref_nodes,df_rest.nodes.unique()):
                            cnt_dst.update({Ld_dic[x,y]:size[y]})





                    e_cnt+=1
                    i+=1


        print ref_time,np.max(df.timestamp)
        assert sum(cnt_dst.values())==n_tot

        if t == t_inf:
            assert DescrStatsW(cnt_dst.keys(),weights=cnt_dst.values()).mean == tot_dic_dst_counter_gen['mean_distance']
            continue

        tot_dic_dst_counter[t] = cnt_dst
        del cnt_dst
        
        assert e_cnt==df.timestamp.unique().shape[0]

        assert t_cnt==e_cnt

        print str(delta_t)+" completed"
    tot_dic_dst_counter_gen['distances'] = tot_dic_dst_counter

    ### Change the folder to statistical analysis if you want to include the
    #create_folder('','Temporal_Network/Other_Networks/statistical_analysis_no_preproc/'+str1)


    #joblib.dump(tot_dic_spk_counter,'/tudelft.net/staff-bulk/ewi/insy/MMC/aceria/Temporal_Network/Other_Networks/statistical_analysis_no_preproc/'+str1+'/spk_tot_stat_'+title1+'.joblib')


    #joblib.dump(tot_dic_dst_counter_gen,'/tudelft.net/staff-bulk/ewi/insy/MMC/aceria/Temporal_Network/Other_Networks/statistical_analysis_no_preproc/'+str1+'/dst_tot_stat_'+title1+'.joblib')



    try: os.mkdir('../Results/avg_top_dist_gvn_temp/data/')
    except: print '../Results/avg_top_dist_gvn_temp/data/ already exist'



            #joblib.dump(tot_dic_spk_counter,'/tudelft.net/staff-bulk/ewi/insy/MMC/aceria/Temporal_Network/Other_Networks/statistical_analysis/'+title+'_rand2_clean/'+'spk_tot_stat_new2_'+title1+'_rand2_'+'seed'+str(seed)+'.joblib')


    joblib.dump(tot_dic_dst_counter_gen,'../Results/avg_top_dist_gvn_temp/data/dst_tot_stat'+title1+'.joblib')

```

```python
def plot_agg_analysis_data(title1):
    
        
    file = joblib.load('../Results/avg_top_dist_gvn_temp/data/dst_tot_stat_'+title1+'.joblib')
    
    return pd.Series({k:DescrStatsW(v.keys(),v.values()).mean/file['mean_distance'] for k,v in file['distances'].items()})


def plot_agg_analysis_random_mean(title1,rand):
    files = [joblib.load(x) for x in glob.glob('../Results/avg_top_dist_gvn_temp/'+title1+'012/'+title1+'_rand*')]
    return pd.concat([pd.Series({k:DescrStatsW(v.keys(),v.values()).mean/x['mean_distance_'+rand] for k,v in x['distances_'+rand].items()}) for x in files],axis=1).mean(axis=1)

def plot_agg_analysis_random_std(title1,rand):
    files = [joblib.load(x) for x in glob.glob('../Results/avg_top_dist_gvn_temp/'+title1+'012/'+title1+'_rand*')]
    return pd.concat([pd.Series({k:DescrStatsW(v.keys(),v.values()).mean/x['mean_distance_'+rand] for k,v in x['distances_'+rand].items()}) for x in files],axis=1).std(axis=1)
```
